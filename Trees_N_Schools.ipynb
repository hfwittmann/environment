# Get List of Towns

```python
import pandas as pd

def getTowns():
    # Start : Global configuration
    # Get Table
    url = "https://de.wikipedia.org/wiki/Liste_der_Gro%C3%9Fst%C3%A4dte_in_Deutschland"
    tables = pd.read_html(url,decimal=',', thousands='.')
    
    grossstadte = tables[0] 
    
    return grossstadte

    
```

```python
townsTable = getTowns()[:-2]
townsTable
```

# Clean Town names

```python
import re

townsTable.loc[:,('Name','Name')] = townsTable.loc[:,('Name','Name')]. \
    apply(lambda x: re.sub(r'[0-9]', '', x)). \
    apply(lambda x:  x.replace('(Oldb)','') ). \
    apply(lambda x: re.sub(r',', '', x)) # removeCommas
townsTable
```

Get Area Ids

```python
from OSMPythonTools.nominatim import Nominatim
nominatim = Nominatim()

def getAreaID(Stadt):
    return nominatim.query(f'{Stadt}, Germany').areaId()
```

Get Number of Schools, Trees

```python
from OSMPythonTools.overpass import overpassQueryBuilder, Overpass
overpass = Overpass()

def getNofObjects(area, selector='"amenity"="school"'):
    # object can be 
    # "amenity"="school"'
    # "natural"="tree"
    query = overpassQueryBuilder(area=int(area), elementType='node', selector=selector, out='count')
    result = overpass.query(query)
    return result.countElements()


def getNofSchools(area):
    # print(area)
    return getNofObjects(area, selector='"amenity"="school"')

def getNofTrees(area):
    return getNofObjects(area, selector='"natural"="tree"')
    
    
```

```python
import math

townsTable.loc[:,('Name','areaID')]=math.nan
townsTable.loc[:,('Name','NoOfSchools')]=math.nan
townsTable.loc[:,('Name','NoOfTrees')]=math.nan
```

```python
for myrow in range(len(townsTable)):
    print(townsTable.loc[myrow,('Name','Name')])
    if not math.isnan(townsTable.loc[myrow,('Name','areaID')]): continue
    
    townsTable.loc[myrow,('Name','areaID')] = getAreaID(townsTable.loc[myrow,('Name','Name')])
```

```python
for myrow in range(len(townsTable)):
    print(townsTable.loc[myrow,('Name','Name')])
    if not math.isnan(townsTable.loc[myrow,('Name','NoOfSchools')]): continue
    
    townsTable.loc[myrow,('Name','NoOfSchools')] = getNofSchools(townsTable.loc[myrow,('Name','areaID')])
```

```python
for myrow in range(len(townsTable)):
    print(townsTable.loc[myrow,('Name','Name')])
    if not math.isnan(townsTable.loc[myrow,('Name','NoOfTrees')]): continue
    
    townsTable.loc[myrow,('Name','NoOfTrees')] = getNofTrees(townsTable.loc[myrow,('Name','areaID')])
```

```python
df = townsTable_extract
df.columns = [pair[1] for pair in df.columns]
df
```

```python
df.columns = ['Name','areaID',	'NoOfSchools',	'NoOfTrees',	'Population', 'Area', 'Population_density']
df
```

```python
for col in ['areaID',	'NoOfSchools',	'NoOfTrees',	'Population', 'Area', 'Population_density']:
    df[col] = df[col].astype('double')
```

```python
df['Trees_density'] = df.NoOfTrees/df.Area
df['Trees_per_person'] = df.NoOfTrees/df.Population
df['Persons_per_tree'] = df.Population / df.NoOfTrees
```

```python
pd.set_option('display.max_rows', 500)
df.sort_values(by=['Trees_density'], ascending=False)
```
